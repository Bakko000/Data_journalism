```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def extract_worldwide_box_office(url):
    # Header per simulare un browser
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    # Effettua la richiesta HTTP
    response = requests.get(url, headers=headers)

    # Controlla se la richiesta è stata eseguita con successo
    if response.status_code == 200:
        # Crea un oggetto BeautifulSoup con il contenuto HTML della pagina
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Trova la tabella contenente i dati
        table = soup.find('div', id='page_filling_chart').find('table')
        
        # Trova tutte le righe della tabella
        rows = table.find_all('tr')
        
        # Itera sulle righe e cerca quella con il testo desiderato
        for row in rows:
            cells = row.find_all('td')
            if len(cells) > 2 and 'Worldwide Box Office' in cells[0].text:
                worldwide_box_office = cells[2].text.strip()
                return worldwide_box_office
    return "Non trovato"

def main(urls, film_names):
    data = []
    for url, film in zip(urls, film_names):
        worldwide_box_office = extract_worldwide_box_office(url)
        data.append({'Film': film, 'Worldwide Box Office': worldwide_box_office})
    
    # Crea un DataFrame pandas con i dati raccolti
    df = pd.DataFrame(data)
    
    # Salva il DataFrame in un file CSV
    df.to_csv('film_worldwide_box_office.csv', index=False)
    print(df)

# Lista di URL e nomi dei film
urls = [
    'https://www.the-numbers.com/movie/Shape-of-Water-The#tab=summary',
    'https://www.the-numbers.com/movie/Forrest-Gump-(1994)#tab=summary'
]

film_names = [
    'The Shape of Water',
    'Forrest Gump'
]

main(urls, film_names)
```
