```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

# Funzione per estrarre il budget da una tabella
def extract_budget_from_table(soup):
    tables = soup.find_all('table')
    target_table = tables[2]  # Poiché la tabella con il budget è la terza (indice 2)
    
    rows = target_table.find_all('tr')
    
    if len(rows) > 3:
        row_3 = rows[3]
        cells = row_3.find_all('td')
        
        cell_data = [cell.get_text(strip=True) for cell in cells]
        budget_text = cell_data[1]  # La seconda cella della riga contiene il budget
        
        match = re.search(r'\$([\d,]+)', budget_text)
        if match:
            budget_number = match.group(1)
            return budget_number
    return None

# Lista di URL e nomi dei film
film_data = [
    {"url": "https://www.the-numbers.com/movie/Shape-of-Water-The#tab=summary", "name": "Shape of Water"},
    {"url": "https://www.the-numbers.com/movie/Forrest-Gump-(1994)#tab=summary", "name": "Forrest Gump"}
    # Aggiungi altri URL e nomi di film qui
]

# Lista per accumulare i dati
data = []

# Ciclo attraverso ogni URL e nome del film
for film in film_data:
    url = film["url"]
    name = film["name"]
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    
    # Effettua la richiesta HTTP
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        budget = extract_budget_from_table(soup)
        
        if budget:
            data.append({"Film": name, "Production Budget": budget})
        else:
            data.append({"Film": name, "Production Budget": "Not found"})
    else:
        data.append({"Film": name, "Production Budget": "Request failed"})

# Crea un DataFrame con i dati raccolti
df = pd.DataFrame(data)

# Stampa il DataFrame
print(df)
```

```python


from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait

# URL of the webpage to scrape
url = 'https://www.the-numbers.com/movie/Shape-of-Water-The#tab=summary'
driver = webdriver.Chrome()
driver.get(url)




# Find the production budget element
try:
    # Explicitly wait for the metrics table to be present
    wait = WebDriverWait(driver, 10)
    metrics_table = wait.until(EC.presence_of_element_located((By.ID, 'movie_ratings')))

    # Locate the row containing "Production Budget"
    rows = metrics_table.find_elements(By.TAG_NAME, 'tr')
    for row in rows:
        cells = row.find_elements(By.TAG_NAME, 'td')
        if len(cells) > 1 and "Production Budget" in cells[0].text:
            # Extract the budget value from the adjacent cell
            budget_value = cells[1].text.strip()
            print(f'Production Budget: {budget_value}')
            break
    else:
        print('Production Budget not found in the table.')

except Exception as e:
    print(f'Error: {e}')

finally:
    # Close the browser
    driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_worldwide_data(urls, specified_titles):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'
    }

    all_data = []

    for url, specified_title in zip(urls, specified_titles):
        # Fetch the webpage
        page = requests.get(url, headers=headers)
        soup = BeautifulSoup(page.content, 'html.parser')

        # Find all div elements with class 'mojo-performance-summary-table'
        performance_summary_divs = soup.find_all('div', class_='mojo-performance-summary-table')

        worldwide_value = None

        for div in performance_summary_divs:
            h2_element = div.find('h2', class_='a-size-large a-text-bold')
            if h2_element and h2_element.text.strip() == 'All Releases':
                release_divs = div.find_all('div', class_='a-section a-spacing-none')
                
                for release_div in release_divs:
                    label_span = release_div.find('span', class_='a-size-small')
                    if label_span:
                        label = label_span.text.strip()

                        if label == 'Worldwide':
                            money_span = release_div.find('span', class_='money')
                            if money_span:
                                worldwide_value = money_span.text.strip()
                                break  # Stop searching further once found

        if worldwide_value:
            all_data.append({
                'Film': specified_title,
                'Incasso Globale': worldwide_value
            })
        else:
            all_data.append({
                'Title': specified_title,
                'Worldwide Gross': 'Dati non disponibili'
            })

    return all_data

# List of URLs to scrape
urls = [
    "https://www.boxofficemojo.com/title/tt0089755/?ref_=bo_se_r_1",
    "https://www.boxofficemojo.com/title/tt0056172/?ref_=bo_se_r_1",
    "https://www.boxofficemojo.com/title/tt0093389/?ref_=bo_se_r_1",
    "https://www.boxofficemojo.com/title/tt0095953/?ref_=bo_se_r_1",
    "https://www.boxofficemojo.com/title/tt0116209/?ref_=bo_se_r_1",
    "https://www.boxofficemojo.com/title/tt0375679/?ref_=bo_se_r_2",
    "https://www.boxofficemojo.com/title/tt1504320/?ref_=bo_se_r_1",
    "https://www.boxofficemojo.com/title/tt4975722/?ref_=bo_se_r_1",
    "https://www.boxofficemojo.com/title/tt6966692/?ref_=bo_se_r_1"
   
    
    
    
    
]

# List of specified titles corresponding to each URL
specified_titles = [
    "La mia Africa",
    "Lawrence d'Arabia",
    "L'ultimo imperatore",
    "Rain Man - L'uomo della pioggia",
    "Il paziente inglese",
    "Crash: Contatto fisico",
    "Il discorso del re",
    "Moonlight", 
    "Green Book"
    
   
    
]

# Scrape data and create DataFrame
data = scrape_worldwide_data(urls, specified_titles)
df_3 = pd.DataFrame(data)

# Print the DataFrame

```
