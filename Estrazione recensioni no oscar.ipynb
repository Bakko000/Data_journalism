```python
import pickle
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
import time
from bs4 import BeautifulSoup

def scrape_reviews(url):
    # Configura il WebDriver (assicurati di avere il chromedriver nella directory specificata o nel PATH)
    driver_path = 'chromedriver-mac-arm64/chromedriver'
    service = ChromeService(executable_path=driver_path)

    # Opzioni per il WebDriver (facoltativo)
    chrome_options = Options()
    # chrome_options.add_argument('--headless')  # Esegui il browser in modalità headless (facoltativo)
    # chrome_options.add_argument('--disable-gpu')  # Necessario per eseguire il browser in modalità headless in Windows (facoltativo)

    driver = webdriver.Chrome(service=service, options=chrome_options)

    # Apri la pagina web
    driver.get(url)

    # Attendi che la pagina si carichi
    time.sleep(3)#3

    # Loop per cliccare su 'Load More' e caricare ulteriori recensioni
    while True:
        try:
            # Trova il pulsante 'Load More'
            load_more_button = driver.find_element(By.ID, 'load-more-trigger')
            
            # Clicca sul pulsante
            load_more_button.click()
            
            # Attendi un po' di tempo per permettere il caricamento dei nuovi contenuti
            time.sleep(5)#
        except:
            # Se il pulsante non è più trovato, esci dal loop
            break

    # Una volta che tutti i contenuti sono stati caricati, estrai il contenuto della pagina
    content = driver.page_source

    # Usa BeautifulSoup per analizzare il contenuto HTML
    soup = BeautifulSoup(content, 'html.parser')

    # Trova tutti i <div> con la classe "text show-more__control"
    divs = soup.find_all('div', class_='text show-more__control')

    # Estrai e raccogli i contenuti dei <div>
    recensioni = []
    for div in divs:
        recensione = div.get_text(strip=True)
        recensioni.append([recensione])

    # Chiudi il driver
    driver.quit()

    return recensioni

# Lista di URL da cui raccogliere le recensioni
urls = [
 'https://www.imdb.com/title/tt0195685/reviews/?ref_=tt_ov_rt',
'https://www.imdb.com/title/tt0395169/reviews/?ref_=tt_ov_rt',
    'https://www.imdb.com/title/tt1205489/reviews/?ref_=tt_ov_rt',
    'https://www.imdb.com/title/tt3312830/reviews/?ref_=tt_ov_rt',
    
]

# Lista di nomi di file corrispondenti agli URL
file_names = [
   'recensioni_erin_brock.pickle',
    'recensioni_hotel_rwanda.pickle',
    'recensioni_gran_torino.pickle',
    'recensioni_youth.pickle'
    
    # aggiungi altri nomi di file qui
]

# Verifica che il numero di URL corrisponda al numero di nomi di file
if len(urls) != len(file_names):
    raise ValueError("Il numero di URL deve corrispondere al numero di nomi di file.")

# Itera su ciascun URL e nome di file e salva le recensioni in file pickle
for url, file_name in zip(urls, file_names):
    recensioni = scrape_reviews(url)
    
    with open(file_name, 'wb') as file:
        pickle.dump(recensioni, file)
    print(f"Recensioni salvate in {file_name}", len(recensioni))

```
