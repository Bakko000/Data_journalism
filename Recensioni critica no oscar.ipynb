```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pickle

# Funzione per caricare e salvare le recensioni da una pagina
def scrape_reviews(url, pickle_filename):
    # Setup Selenium WebDriver (assuming Chrome for this example)
    driver = webdriver.Chrome()

    try:
        # Navigate to the page with the reviews
        driver.get(url)

        # Trova il bottone "Load More" e cliccalo finché è visibile
        while True:
            try:
                load_more_button = driver.find_element(By.XPATH, "//rt-button[contains(text(), 'Load More')]")
                if load_more_button.is_displayed():
                    load_more_button.click()
                    time.sleep(3)  # Aspetta che le nuove recensioni vengano caricate (puoi regolare il tempo in base alla velocità del caricamento)
                else:
                    break
            except:
                break

        # Trova tutti gli elementi del testo della recensione
        review_quotes = driver.find_elements(By.CLASS_NAME, "review-text")

        # Inizializza una lista per salvare le recensioni
        reviews_list = []

        # Estrai ogni citazione della recensione e aggiungila alla lista
        for quote in review_quotes:
            reviews_list.append([quote.text])

        # Salva le recensioni in un file pickle
        with open(pickle_filename, 'wb') as f:
            pickle.dump(reviews_list, f)
            print(len(reviews_list))

    finally:
        # Chiudi il browser alla fine, anche se ci sono stati errori
        driver.quit()

# Lista di URL delle pagine da cui estrarre le recensioni
urls = [

  'https://www.rottentomatoes.com/m/erin_brockovich/reviews?type=top_critics',
    'https://www.rottentomatoes.com/m/hotel_rwanda/reviews?type=top_critics',
    'https://www.rottentomatoes.com/m/gran_torino/reviews?type=top_critics',
    'https://www.rottentomatoes.com/m/youth_2015/reviews?type=top_critics'
    
    # aggiungi altri URL qui
]

# Lista di nomi dei file pickle corrispondenti
pickle_filenames = [
   
    'recensioni_critica_erin_brock.pickle',
    'recensioni_critica_hotel_rwanda.pickle',
    'recensioni_critica_gran_torino.pickle',
    'recensioni_critica_youth.pickle'
    
    # aggiungi altri nomi di file qui
]

# Verifica che il numero di URL e nomi dei file pickle corrisponda
if len(urls) != len(pickle_filenames):
    raise ValueError("Il numero di URL e nomi dei file pickle deve essere lo stesso")

# Itera attraverso ogni URL e chiama la funzione scrape_reviews
for url, pickle_filename in zip(urls, pickle_filenames):
    scrape_reviews(url, pickle_filename)
    print(f"Recensioni salvate in {pickle_filename}")
```
